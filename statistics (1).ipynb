### **1. What is a random variable in probability theory?**

A random variable is a function that assigns numerical values to the outcomes of a random experiment. It can represent things like the result of rolling a die or measuring temperature.

---

### **2. What are the types of random variables?**

There are two types:

* **Discrete Random Variable:** Takes specific, countable values (e.g., number of students).
* **Continuous Random Variable:** Takes any value within a range (e.g., height, weight).

---

### **3. What is the difference between discrete and continuous distributions?**

* **Discrete distribution:** Deals with variables that take distinct values. Probabilities are assigned to each value.
* **Continuous distribution:** Deals with variables over an interval. Probabilities are given over ranges using an area under the curve.

---

### **4. What are probability distribution functions (PDF)?**

* For **discrete variables**, it's called the **Probability Mass Function (PMF)**.
* For **continuous variables**, it's called the **Probability Density Function (PDF)**.
  These describe how probabilities are distributed over the values of the random variable.

---

### **5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

* **PDF:** Gives probability for exact values (in discrete) or probability density (in continuous).
* **CDF:** Gives the cumulative probability that the variable takes a value less than or equal to a specific value:
  $F(x) = P(X \leq x)$

---

### **6. What is a discrete uniform distribution?**

A distribution where each outcome has an equal probability.
Example: Rolling a fair 6-sided die — each number (1–6) has a probability of 1/6.

---

### **7. What are the key properties of a Bernoulli distribution?**

* Only two outcomes: Success (1) and Failure (0)
* Probability of success = **p**, failure = **1 − p**
* Mean = **p**, Variance = **p(1 − p)**

---

### **8. What is the binomial distribution, and how is it used in probability?**

It models the number of successes in **n independent Bernoulli trials** with success probability **p**.
Used in scenarios like counting correct answers in a quiz.

---

### **9. What is the Poisson distribution and where is it applied?**

Models the number of times an event occurs in a fixed interval of time or space, when events occur independently.
Used for modeling things like call arrivals or traffic incidents.

---

### **10. What is a continuous uniform distribution?**

A distribution where every value in a continuous interval $[a, b]$ has the same probability density.
PDF:
$f(x) = \frac{1}{b-a} \text{ for } a \leq x \leq b$

---

### **11. What are the characteristics of a normal distribution?**

* Bell-shaped, symmetric curve
* Mean = Median = Mode
* Defined by mean (μ) and standard deviation (σ)
* 68% of data within 1σ, 95% within 2σ, 99.7% within 3σ

---

### **12. What is the standard normal distribution, and why is it important?**

A normal distribution with:

* Mean = 0
* Standard deviation = 1
  It is used to calculate probabilities using **Z-scores**, and to standardize data.

---

### **13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

CLT states that the **sampling distribution of the sample mean** becomes approximately normal as the sample size increases (n ≥ 30), regardless of the population’s distribution.
It's fundamental for conducting hypothesis testing and confidence intervals.

---

### **14. How does the Central Limit Theorem relate to the normal distribution?**

It justifies using the **normal distribution** to approximate the distribution of sample means, even if the population is not normally distributed.

---

### **15. What is the application of Z statistics in hypothesis testing?**

Z-statistics are used to test hypotheses about population means when the population standard deviation is known. It helps determine how far a sample mean deviates from the population mean.

---

### **16. How do you calculate a Z-score, and what does it represent?**

**Formula:**
$Z = \frac{X - \mu}{\sigma}$
It shows how many standard deviations a data point **X** is from the mean **μ**.

---

### **17. What are point estimates and interval estimates in statistics?**

* **Point Estimate:** A single value estimate of a population parameter (e.g., sample mean).
* **Interval Estimate:** A range of values (e.g., confidence interval) likely to contain the population parameter.

---

### **18. What is the significance of confidence intervals in statistical analysis?**

They provide a range where the true population parameter is expected to fall, with a certain level of confidence (e.g., 95%).

---

### **19. What is the relationship between a Z-score and a confidence interval?**

Z-scores are used to calculate the bounds of a confidence interval:
$CI = \bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$

---

### **20. How are Z-scores used to compare different distributions?**

Z-scores standardize values, enabling comparison across datasets with different means and standard deviations.

---

### **21. What are the assumptions for applying the Central Limit Theorem?**

* Sample is random
* Observations are independent
* Sample size is large (typically n ≥ 30)
* Population has finite variance

---

### **22. What is the concept of expected value in a probability distribution?**

Expected value is the **long-term average** value of a random variable:
For discrete:
$E(X) = \sum x_i P(x_i)$
For continuous:
$E(X) = \int x f(x) dx$

---

### **23. How does a probability distribution relate to the expected outcome of a random variable?**

The expected value is calculated using the probability distribution. It reflects the **average outcome** you would expect after many repetitions of the experiment.
